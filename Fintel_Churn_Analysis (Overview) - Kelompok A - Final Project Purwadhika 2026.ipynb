> **Definisi *Churn***

Variabel ***Churn*** merupakan variabel target pada dataset yang menunjukkan apakah seorang pelanggan akan berhenti berlangganan dalam satu bulan ke depan. Variabel ini digunakan sebagai label (*target variable*) dalam model *machine learning* untuk memprediksi *customer churn* berdasarkan pola historis pelanggan.

Adapun nilai pada variabel **Churn** dalam *machine learning* diinterpretasikan sebagai berikut:

- ***Yes (1)*** — pelanggan akan berhenti berlangganan (*churn*) dalam satu bulan ke depan.  
- ***No (0)*** — pelanggan tidak akan berhenti berlangganan (*non-churn*) dalam satu bulan ke depan.

> **Pemilihan Metrik berdasarkan Tujuan dan *Stakeholder*:**

**Metrik Utama: F-beta *Score* dengan β = 2**

Pemilihan metrik evaluasi didasarkan pada asimetri biaya yang signifikan antara dua jenis kesalahan prediksi:

| **Jenis Kesalahan** | **Dampak Bisnis** | **Estimasi Biaya** |
|:---|:---|---:|
| **False Negative (FN)** — *churner* tidak terdeteksi | Pelanggan benar-benar berhenti berlangganan sehingga perusahaan kehilangan pendapatan dan harus mengeluarkan biaya untuk memperoleh pelanggan pengganti. | **≈ \$42.20** per pelanggan *(estimasi biaya akuisisi pelanggan baru)* |
| **False Positive (FP)** — *non-churner* diprediksi sebagai *churn* | Perusahaan memberikan program retensi kepada pelanggan yang sebenarnya tidak berisiko, sehingga menimbulkan biaya intervensi yang tidak diperlukan. | **≈ \$10.55** per pelanggan *(estimasi biaya diskon/voucher retensi)* |

Karena biaya akuisisi pelanggan baru sekitar **4× lebih tinggi** dibandingkan biaya retensi (Junxiang Lu, 2002 dalam Yulianti & Saifudin, 2020), maka setiap *False Negative* dianggap jauh lebih merugikan dibanding *False Positive*.

Oleh karena itu, digunakan **F-beta *Score* dengan β = 2**, yang memberikan bobot lebih besar pada *recall* (4× lebih penting) dibanding *precision* , sehingga model lebih sensitif dalam menangkap pelanggan yang berpotensi *churn*.

$$F_\beta = (1 + \beta^2) \cdot \frac{\text{Precision} \times \text{Recall}}{\beta^2 \cdot \text{Precision} + \text{Recall}}, \quad \beta = 2$$


**Metrik Sekunder:** **ROC-AUC**, digunakan untuk mengevaluasi kemampuan model dalam membedakan pelanggan *churn* dan *non-churn* berdasarkan skor probabilitas, serta sebagai dasar dalam proses *threshold tuning*.
